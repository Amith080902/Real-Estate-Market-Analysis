```python
import pandas as pd

# Load the dataset
df = pd.read_csv("C:/Users/AMIT/OneDrive/Desktop/realtor-data.zip.csv")
```

```python
# Display the first few rows
print(df.head())

# Get a summary of the dataset
print(df.info())

# Check for missing values
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of property prices
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Average price per state
avg_price_state = df.groupby('state')['price'].mean().sort_values()
avg_price_state.plot(kind='barh', figsize=(10,8))
plt.title('Average Property Price by State')
plt.xlabel('Average Price')
plt.ylabel('State')
plt.show()
```

```python
# Save the cleaned data to a new CSV file
df.to_csv('Cleaned_Real_Estate_Data.csv', index=False)
```
